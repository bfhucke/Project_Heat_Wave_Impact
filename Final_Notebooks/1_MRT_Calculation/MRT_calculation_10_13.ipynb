```python
# import needed libraries
import numpy as np
import pandas as pd
import xarray as xr
```

```python
ds = xr.open_dataset('C:/Users/benhu/MasterThesisRawData/download10_13_rad.nc')
rad_13 = ds.to_dataframe()
rad_13
```

```python
# include latitude and logitude as column and round numbers
rad_13 = rad_13.reset_index(level=['longitude', 'latitude', 'time'])
rad_13
```

```python
# convert from J mÂ² to W mÂ²
rad_13['ssr'] = rad_13['ssr'] / 3600
rad_13['str'] = rad_13['str'] / 3600
rad_13['ssrd'] = rad_13['ssrd'] / 3600
rad_13['strd'] = rad_13['strd'] / 3600
rad_13['fdir'] = rad_13['fdir'] / 3600
```

```python
# load obtained locations for the coordinates
locations = pd.read_csv('locations1.csv', index_col=0)
# merging locations with weather data
rad_13 = pd.merge(rad_13, locations,  how='left', left_on=['latitude','longitude'], right_on = ['lat','lon'])
```

```python
rad_13.shape
```

```python
# drop observations that are not assigned to a country (probably coordinates on water)
rad_13 = rad_13.dropna(subset=['country'])
rad_13.shape
```

```python
rad_13
```

```python
# computing first components for MRT formula (equation number according to equation numbers in Di Napoli et al., 2020)
# upwelling thermal component (equation 3)
rad_13['stru'] = rad_13['strd']-rad_13['str']
# compute diffuse solar radiation flux (equation 4)
rad_13['fdif'] = rad_13['ssrd']-rad_13['fdir']
# compute surface-reflected solar radiation flux (equation 5)
rad_13['ssru'] = rad_13['ssrd']-rad_13['ssr']
```

```python
# convert time to datetime
rad_13['time'] = pd.to_datetime(rad_13['time'])
```

```python
# get day, month, year and hour from datetime object
rad_13['day'] = rad_13.time.dt.day
rad_13['month'] = rad_13.time.dt.month
rad_13['year'] = rad_13.time.dt.year
rad_13['hour'] = rad_13.time.dt.hour
```

```python
# get Day of Year from date
rad_13['date'] = pd.to_datetime(rad_13['time']).dt.date
rad_13['date'] = pd.to_datetime(rad_13['date'])
rad_13['DOY'] = rad_13['date'].dt.dayofyear
```

```python
# formula sun declination: https://www.pveducation.org/pvcdrom/properties-of-sunlight/declination-angle#footnote1_soeo4fz
rad_13['sda'] = -23.45*np.cos(np.radians(360/365*(rad_13['DOY']+10)))
```

```python
# sunrise (multiplied by pi over 180 to get radians)
rad_13['sunrise'] = -np.tan(rad_13['latitude']*(np.pi/180))*np.tan(rad_13['sda']*(np.pi/180))
rad_13['sunrise'] = np.arccos(rad_13['sunrise'])
rad_13['sunrise'] = np.degrees(-rad_13['sunrise'])

#sunset
rad_13['sunset'] = -np.tan(rad_13['latitude']*(np.pi/180))*np.tan(rad_13['sda']*(np.pi/180))
rad_13['sunset'] = np.arccos(rad_13['sunset'])
rad_13['sunset'] = np.degrees(rad_13['sunset'])
```

```python
# reading UTC per NUTS thats obtained from the Notebook for 2000-2001
NUTS_time_zone = pd.read_csv('NUTS_time_zone.csv')
rad_13 = rad_13.merge(NUTS_time_zone, how='inner', on=['NUTS1'])
rad_13.head()
```

```python
rad_13['time_zone_UTC'].value_counts()
```

```python
rad_13.head(3)
```

```python
# calculate hour angle at given hour (https://solarsena.com/solar-hour-angle-calculator-formula/)
# STEP 1: get fractional year in radians
rad_13['year_fraction'] = ((2*np.pi)/365)*(rad_13['DOY']-1+((rad_13['hour']-12)/24))
# STEP 2: use equation of time
rad_13['EOT'] = 229.18*(0.000075+0.001868*np.cos(rad_13['year_fraction'])
                    -0.032077*np.sin(rad_13['year_fraction'])
                    -0.014615*np.cos(2*rad_13['year_fraction'])
                    -0.040849*np.sin(2*rad_13['year_fraction']))
# STEP 3: compute offset
rad_13['offset'] = rad_13['EOT']+4*(rad_13['longitude']-15*rad_13['time_zone_UTC'])
# STEP 4: compute corrected local solar time
rad_13['hour_corrected'] = rad_13['hour']+rad_13['offset']/60
# STEP 5: get solar hour angle
rad_13['sha'] = 15*(rad_13['hour_corrected']-12)
```

```python
# get solar zenith angle (https://en.wikipedia.org/wiki/Solar_zenith_angle)
rad_13['z_angle'] = np.sin(np.radians(rad_13['latitude']))*np.sin(np.radians(rad_13['sda']))+(np.cos(np.radians(rad_13['latitude']))*np.cos(np.radians(rad_13['sda']))*np.cos(np.radians(rad_13['sha'])))
rad_13['z_angle'] = np.arccos(rad_13['z_angle'])
rad_13['z_angle'] = np.degrees(rad_13['z_angle'])
rad_13.sort_values(by='z_angle')
```

```python
# get solar elevation angle (equation 16)
rad_13['elevation'] = 90 - rad_13['z_angle']
```

```python
# get surface projection factor (equation 15)
a = rad_13['elevation'] * (0.998-rad_13['elevation']**2/50000)
rad_13['f_p'] = 0.308*np.cos(a)
rad_13.head()
```

```python
# insepct observations with NaN in sunrise
pd.set_option('display.max_columns', None)
gg = rad_13[rad_13['sunrise'].isnull()]
gg.shape
```

Only observations above the arctic circle have missing values in the sunrise and sunset hour angle columns between June 10 and July 2. This is because in these places the sun does not set in this time period. The NaN will be set to 0.

```python
# fill NaN with the maximum possible value
rad_13['sunrise'] = rad_13['sunrise'].fillna(-179.99)
rad_13['sunset'] = rad_13['sunset'].fillna(179.99)
rad_13[rad_13['sunrise'].isnull()]
```

```python
# testing whether NaN were filled correctly
o = rad_13[(rad_13['latitude'] == 68) & (rad_13['month'] == 6) & (rad_13['day'] == 21)]
o.head()
```

```python
# compute average daytime cosine of solar zenith angle (equation 12)
rad_13['cos_teta'] = np.sin(np.radians(rad_13['sda']))*np.sin(np.radians(rad_13['latitude']))+(1/rad_13['sunset']-rad_13['sunrise'])*np.cos(np.radians(rad_13['sda']))*np.cos(np.radians(rad_13['latitude']))*(np.sin(np.radians(rad_13['sunset']))-np.sin(np.radians(rad_13['sunrise'])))
```

```python
# get direct component I*
rad_13['I_asterisk'] = rad_13['fdir']/rad_13['cos_teta']
```

```python
from scipy.constants import Stefan_Boltzmann
```

```python
Stefan_Boltzmann
```

```python
# get MRT (equation 14)
rad_13['MRT'] = 1/Stefan_Boltzmann*(0.5*rad_13['strd']+0.5*rad_13['stru']+(0.7/0.97)*(0.5*rad_13['fdif']+0.5*rad_13['ssru']+rad_13['f_p']*rad_13['I_asterisk']))
rad_13['MRT'] = rad_13['MRT']**0.25
# get Celsius
rad_13['MRT'] = rad_13['MRT'] - 273.15
```

# Convert Data from hourly to daily

```python
# sub frames per hour
sub2am = MRT[MRT['hour'] == 2]
sub11am = MRT[MRT['hour'] == 11]
sub16am = MRT[MRT['hour'] == 16]
sub23am = MRT[MRT['hour'] == 23]
```

```python
# rename columns by hour
sub2am = sub2am.rename(columns={"MRT": "MRT_2AM"})
sub11am = sub11am.rename(columns={"MRT": "MRT_11AM"})
sub16am = sub16am.rename(columns={"MRT": "MRT_4PM"})
sub23am = sub23am.rename(columns={"MRT": "MRT_11PM"})
```

```python
# dropping time and hour variables
sub2am = sub2am.drop(['hour','time'],1)
sub11am = sub11am.drop(['hour','time'],1)
sub16am = sub16am.drop(['hour','time'],1)
sub23am = sub23am.drop(['hour','time'],1)
```

```python
# merging subsets of data per hour
df1 = sub2am.merge(sub11am, how='inner', on=['date','latitude', 'longitude'])
df2 = df1.merge(sub16am, how='inner', on=['date','latitude', 'longitude'])
final02_13 = df2.merge(sub23am, how='inner', on=['date','latitude', 'longitude'])
```

```python
# merge with other variables
final_rad_13 = t.merge(final02_13, how='inner', on=['date','latitude', 'longitude'])
```
