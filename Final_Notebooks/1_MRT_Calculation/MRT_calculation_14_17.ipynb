```python
# import needed libraries
import numpy as np
import pandas as pd
import xarray as xr
```

```python
ds = xr.open_dataset('C:/Users/benhu/MasterThesisRawData/download14_17_rad.nc')
rad_17 = ds.to_dataframe()
rad_17
```

```python
# include latitude and logitude as column and round numbers
rad_17 = rad_17.reset_index(level=['longitude', 'latitude', 'time'])
rad_17
```

```python
# convert from J m² to W m²
rad_17['ssr'] = rad_17['ssr'] / 3600
rad_17['str'] = rad_17['str'] / 3600
rad_17['ssrd'] = rad_17['ssrd'] / 3600
rad_17['strd'] = rad_17['strd'] / 3600
rad_17['fdir'] = rad_17['fdir'] / 3600
```

```python
# load obtained locations for the coordinates
locations = pd.read_csv('locations1.csv', index_col=0)
# merging locations with weather data
rad_17 = pd.merge(rad_17, locations,  how='left', left_on=['latitude','longitude'], right_on = ['lat','lon'])
```

```python
rad_17.shape
```

```python
# drop observations that are not assigned to a country (probably coordinates on water)
rad_17 = rad_17.dropna(subset=['country'])
rad_17.shape
```

```python
rad_17
```

```python
# computing first components for MRT formula (equation number according to equation numbers in Di Napoli et al., 2020)
# upwelling thermal component (equation 3)
rad_17['stru'] = rad_17['strd']-rad_17['str']
# compute diffuse solar radiation flux (equation 4)
rad_17['fdif'] = rad_17['ssrd']-rad_17['fdir']
# compute surface-reflected solar radiation flux (equation 5)
rad_17['ssru'] = rad_17['ssrd']-rad_17['ssr']
```

```python
# convert time to datetime
rad_17['time'] = pd.to_datetime(rad_17['time'])
```

```python
# get day, month, year and hour from datetime object
rad_17['day'] = rad_17.time.dt.day
rad_17['month'] = rad_17.time.dt.month
rad_17['year'] = rad_17.time.dt.year
rad_17['hour'] = rad_17.time.dt.hour
```

```python
# get Day of Year from date
rad_17['date'] = pd.to_datetime(rad_17['time']).dt.date
rad_17['date'] = pd.to_datetime(rad_17['date'])
rad_17['DOY'] = rad_17['date'].dt.dayofyear
```

```python
# formula sun declination: https://www.pveducation.org/pvcdrom/properties-of-sunlight/declination-angle#footnote1_soeo4fz
rad_17['sda'] = -23.45*np.cos(np.radians(360/365*(rad_17['DOY']+10)))
```

```python
# sunrise (multiplied by pi over 180 to get radians)
rad_17['sunrise'] = -np.tan(rad_17['latitude']*(np.pi/180))*np.tan(rad_17['sda']*(np.pi/180))
rad_17['sunrise'] = np.arccos(rad_17['sunrise'])
rad_17['sunrise'] = np.degrees(-rad_17['sunrise'])

#sunset
rad_17['sunset'] = -np.tan(rad_17['latitude']*(np.pi/180))*np.tan(rad_17['sda']*(np.pi/180))
rad_17['sunset'] = np.arccos(rad_17['sunset'])
rad_17['sunset'] = np.degrees(rad_17['sunset'])
```

```python
# reading UTC per NUTS thats obtained from the Notebook for 2000-2001
NUTS_time_zone = pd.read_csv('NUTS_time_zone.csv')
rad_17 = rad_17.merge(NUTS_time_zone, how='inner', on=['NUTS1'])
rad_17.head()
```

```python
rad_17.head(3)
```

```python
# calculate hour angle at given hour (https://solarsena.com/solar-hour-angle-calculator-formula/)
# STEP 1: get fractional year in radians
rad_17['year_fraction'] = ((2*np.pi)/365)*(rad_17['DOY']-1+((rad_17['hour']-12)/24))
# STEP 2: use equation of time
rad_17['EOT'] = 229.18*(0.000075+0.001868*np.cos(rad_17['year_fraction'])
                    -0.032077*np.sin(rad_17['year_fraction'])
                    -0.014615*np.cos(2*rad_17['year_fraction'])
                    -0.040849*np.sin(2*rad_17['year_fraction']))
# STEP 3: compute offset
rad_17['offset'] = rad_17['EOT']+4*(rad_17['longitude']-15*rad_17['time_zone_UTC'])
# STEP 4: compute corrected local solar time
rad_17['hour_corrected'] = rad_17['hour']+rad_17['offset']/60
# STEP 5: get solar hour angle
rad_17['sha'] = 15*(rad_17['hour_corrected']-12)
```

```python
# get solar zenith angle (https://en.wikipedia.org/wiki/Solar_zenith_angle)
rad_17['z_angle'] = np.sin(np.radians(rad_17['latitude']))*np.sin(np.radians(rad_17['sda']))+(np.cos(np.radians(rad_17['latitude']))*np.cos(np.radians(rad_17['sda']))*np.cos(np.radians(rad_17['sha'])))
rad_17['z_angle'] = np.arccos(rad_17['z_angle'])
rad_17['z_angle'] = np.degrees(rad_17['z_angle'])
rad_17.sort_values(by='z_angle')
```

```python
# get solar elevation angle (equation 16)
rad_17['elevation'] = 90 - rad_17['z_angle']
```

```python
# get surface projection factor (equation 15)
a = rad_17['elevation'] * (0.998-rad_17['elevation']**2/50000)
rad_17['f_p'] = 0.308*np.cos(a)
rad_17.head()
```

```python
# insepct observations with NaN in sunrise
pd.set_option('display.max_columns', None)
gg = rad_17[rad_17['sunrise'].isnull()]
gg.shape
```

Only observations above the arctic circle have missing values in the sunrise and sunset hour angle columns between June 10 and July 2. This is because in these places the sun does not set in this time period. The NaN will be set to 0.

```python
# fill NaN with the maximum possible value
rad_17['sunrise'] = rad_17['sunrise'].fillna(-179.99)
rad_17['sunset'] = rad_17['sunset'].fillna(179.99)
rad_17[rad_17['sunrise'].isnull()]
```

```python
# testing whether NaN were filled correctly
o = rad_17[(rad_17['latitude'] == 68) & (rad_17['month'] == 6) & (rad_17['day'] == 21)]
o.head()
```

```python
# compute average daytime cosine of solar zenith angle (equation 12)
rad_17['cos_teta'] = np.sin(np.radians(rad_17['sda']))*np.sin(np.radians(rad_17['latitude']))+(1/rad_17['sunset']-rad_17['sunrise'])*np.cos(np.radians(rad_17['sda']))*np.cos(np.radians(rad_17['latitude']))*(np.sin(np.radians(rad_17['sunset']))-np.sin(np.radians(rad_17['sunrise'])))
```

```python
# get direct component I*
rad_17['I_asterisk'] = rad_17['fdir']/rad_17['cos_teta']
```

```python
from scipy.constants import Stefan_Boltzmann
```

```python
Stefan_Boltzmann
```

```python
# get MRT (equation 14)
rad_17['MRT'] = 1/Stefan_Boltzmann*(0.5*rad_17['strd']+0.5*rad_17['stru']+(0.7/0.97)*(0.5*rad_17['fdif']+0.5*rad_17['ssru']+rad_17['f_p']*rad_17['I_asterisk']))
rad_17['MRT'] = rad_17['MRT']**0.25
# get Celsius
rad_17['MRT'] = rad_17['MRT'] - 273.15
```

# Convert Data from hourly to daily

```python
# sub frames per hour
sub2am = MRT[MRT['hour'] == 2]
sub11am = MRT[MRT['hour'] == 11]
sub16am = MRT[MRT['hour'] == 16]
sub23am = MRT[MRT['hour'] == 23]
```

```python
# rename columns by hour
sub2am = sub2am.rename(columns={"MRT": "MRT_2AM"})
sub11am = sub11am.rename(columns={"MRT": "MRT_11AM"})
sub16am = sub16am.rename(columns={"MRT": "MRT_4PM"})
sub23am = sub23am.rename(columns={"MRT": "MRT_11PM"})
```

```python
# dropping time and hour variables
sub2am = sub2am.drop(['hour','time'],1)
sub11am = sub11am.drop(['hour','time'],1)
sub16am = sub16am.drop(['hour','time'],1)
sub23am = sub23am.drop(['hour','time'],1)
```

```python
# merging subsets of data per hour
df1 = sub2am.merge(sub11am, how='inner', on=['date','latitude', 'longitude'])
df2 = df1.merge(sub16am, how='inner', on=['date','latitude', 'longitude'])
final02_17 = df2.merge(sub23am, how='inner', on=['date','latitude', 'longitude'])
```

```python
# merge with other variables
final_rad_17 = t.merge(final02_17, how='inner', on=['date','latitude', 'longitude'])
```
